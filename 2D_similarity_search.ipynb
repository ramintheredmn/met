```python
from rdkit import Chem
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from rdkit import Chem, DataStructs, IPythonConsole
from rdkit.Chem import (
    PandasTools,
    Draw,
    Descriptors,
    MACCSkeys,
    rdFingerprintGenerator,
)
IPythonConsole.ipython_useSVG = True

from tqdm.auto import tqdm
```

```python
tqdm.pandas()
```

```python
all_fda_df = pd.read_csv("../all_fda_approved.csv", engine='pyarrow', dtype_backend='pyarrow')
```

```python
all_fda_df.head()
```

```python
all_fda_df['mol'] = all_fda_df.SMILES.progress_apply(Chem.MolFromSmiles)
```

```python
all_fda_df.head()
```

```python
Chem.Draw.MolsToGridImage(
    all_fda_df.mol.values[0:10],
    legends=all_fda_df.name.values[0:10]
    
)
```

```python
all_fda_df['maccs'] = all_fda_df.mol.apply(MACCSkeys.GenMACCSKeys)
```

```python
all_fda_df.head()
```

```python
query = Chem.MolFromSmiles("C[C@@H](c1ccc2nccn2c1)n1nnc2ncc(-c3cnn(C)c3)nc21")
```

```python
query_maccs = MACCSkeys.GenMACCSKeys(query)
```

```python
all_fda_df['tanimoto_maccs'] = DataStructs.BulkTanimotoSimilarity(query_maccs, all_fda_df.maccs)
```

```python
all_fda_df.sort_values('tanimoto_maccs', ascending=False).head(10)
```

```python
Chem.Draw.MolsToGridImage(
    [query, *all_fda_df.sort_values('tanimoto_maccs', ascending=False).head(30).mol.values],
    legends = ['query', *all_fda_df.sort_values('tanimoto_maccs', ascending=False).head(30).name.values]
)
```

```python
fingerprintgen = rdFingerprintGenerator.GetMorganGenerator(radius=2, fpSize=1024)
```

```python
all_fda_df['morgan'] = all_fda_df.mol.map(fingerprintgen.GetFingerprint)
```

```python
all_fda_df.head()
```

```python
query_moragan = fingerprintgen.GetFingerprint(query)
```

```python
all_fda_df['tanimoto_morgan'] = DataStructs.BulkTanimotoSimilarity(query_moragan, all_fda_df.morgan.values)
```

```python
all_fda_df.sort_values('tanimoto_morgan', ascending=False).head(10)
```

```python
Chem.Draw.MolsToGridImage(
    [query, *all_fda_df.sort_values('tanimoto_morgan', ascending=False).head(10).mol.values],
    legends = ['query', *all_fda_df.sort_values('tanimoto_morgan', ascending=False).head(10).name.values]
)
```

```python
all_fda_df['dice_maccs'] = DataStructs.BulkDiceSimilarity(query_maccs, all_fda_df.maccs.values)
```

```python
all_fda_df['dice_morgan'] = DataStructs.BulkDiceSimilarity(query_moragan, all_fda_df.morgan.values)
```

```python
all_fda_df.head()
```

```python
tanimoto_maccs = (all_fda_df
 .query("tanimoto_maccs >= 0.4")
 .sort_values('tanimoto_maccs', ascending=False)
 [['molecule_chembl_id', 'SMILES', 'name', 'tanimoto_maccs']]
 .rename(columns = {'tanimoto_maccs': 'score'})
 .assign(
     method = 'tanimoto_maccs'
 )
        
)
```

```python
tanimoto_morgan = (all_fda_df
 .query("tanimoto_morgan >= 0.15")
 .sort_values('tanimoto_maccs', ascending=False)
 [['molecule_chembl_id', 'SMILES', 'name', 'tanimoto_morgan']]
 .rename(columns = {'tanimoto_morgan': 'score'})
 .assign(
     method = 'tanimoto_morgan'
 )

)
```

```python
dice_maccs = (all_fda_df
 .query("dice_maccs >= 0.4")
 .sort_values('dice_maccs', ascending=False)
 [['molecule_chembl_id', 'SMILES', 'name', 'dice_maccs']]
 .rename(columns = {'dice_maccs': 'score'})
 .assign(
     method = 'dice_maccs'
 )
)
```

```python
dice_morgan = (all_fda_df
 .query("dice_morgan >= 0.2")
 .sort_values('dice_morgan', ascending=False)
 [['molecule_chembl_id', 'SMILES', 'name', 'dice_morgan']]
 .rename(columns = {'dice_morgan': 'score'})
 .assign(
     method = 'dice_morgan'
 )
)
```

```python
swiwss_similar_df = pd.read_csv("./savolotinib_6SDE/swiss_similar.csv", engine='pyarrow', dtype_backend='pyarrow')
```

```python
swiss = (swiwss_similar_df
 .query("Score >= 0.3")
 [['Molecule_ID', 'SMILES', 'Score', 'pref_name']]
 .rename(columns = {'Molecule_ID':'molecule_chembl_id', 'Score': 'score', 'pref_name':'name'})
 .assign(
     method = 'swiss'
 )
)
```

```python
swiss.head()
```

```python
all_ = pd.concat([
    swiss,
    dice_morgan,
    tanimoto_morgan,
    dice_maccs,
    tanimoto_maccs
    
], axis=0)
```
