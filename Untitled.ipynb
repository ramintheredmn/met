```python
import pandas as pd
from rdkit import Chem, DataStructs, IPythonConsole
from rdkit.Chem import (
    PandasTools,
    Draw,
    Descriptors,
    MACCSkeys,
    rdFingerprintGenerator,
)
IPythonConsole.ipython_useSVG = True
```

```python
df_final = pd.read_csv("./savolotinib_6SDE_lgbm/final_data.csv")
```

```python
df_final.head()
```

```python
df_final['mol'] = df_final.SMILES.apply(Chem.MolFromSmiles)
```

```python
fingerprintgen = rdFingerprintGenerator.GetMorganGenerator(radius=4, fpSize=2048)
```

```python
df_final['fn'] = df_final.mol.map(MACCSkeys.GenMACCSKeys)
```

## Capmatinib

```python
capmatinib_mol = Chem.MolFromSmiles("CNC(=O)c1ccc(-c2cnc3ncc(Cc4ccc5ncccc5c4)n3n2)cc1F")
capmatinib_mol
```

```python
query = MACCSkeys.GenMACCSKeys(capmatinib_mol)
```

```python
df_final['capmatinib_tanimoto'] = DataStructs.BulkTanimotoSimilarity(query, df_final.fn.values)
```

## Crizotinib

```python
crizotinib_mol = Chem.MolFromSmiles("C1(=C(C=CC(=C1[C@H](OC2=C(N=CC(=C2)C3=C[N](N=C3)C4CCNCC4)N)C)Cl)F)Cl")
crizotinib_mol
```

```python
query = MACCSkeys.GenMACCSKeys(crizotinib_mol)
```

```python
df_final['crizotinib_tanimoto'] = DataStructs.BulkTanimotoSimilarity(query, df_final.fn.values)
```

## Savolotinib

```python
savolitinib_mol = Chem.MolFromSmiles("C[C@@H](c1ccc2nccn2c1)n1nnc2ncc(-c3cnn(C)c3)nc21")
```

```python
query = MACCSkeys.GenMACCSKeys(savolitinib_mol)
```

```python
df_final['savolotinib_tanimoto'] = DataStructs.BulkTanimotoSimilarity(query, df_final.fn.values)
```

```python
df_final.drop(columns=['fn', 'morgan']).query("crizotinib_tanimoto > 0.5 or savolotinib_tanimoto > 0.5")
```

```python
df_final.drop(columns=['fn', 'morgan']).to_csv("tanimoto_result.csv", index=False)
```
